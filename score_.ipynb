Lexical diversity: the average number of times a given word has repeated across the text. Simply put, lexical diversity is (total # of words in text / # of unique words in text)

```
def lexical_diversity(text):
    try:
        return len(set(text)) / len(text) 
    except:
        return 1
```

```
lexical_diversity(contents.split(' '))
```

Compute the Euclidean distance for one dimension. The distance between two points in one dimension is simply the absolute value of the difference between their coordinates. 

If the points A(x1,y1) and B(x2,y2) are in 2-dimensional space, then the Euclidean distance between them is

|AB| = √ ((x2-x1)^2 + (y2-y1)^2)

If the points A(x1,y1,z1) and B(x2,y2,z2) are in 3-dimensional space, then the Euclidean distance between them is

 |AB| = √ ((x2-x1)^2 +(y2-y1)^2 +(z2-z1)^2)

```
#distance btw keywords and content
import numpy as np


dist=np.sqrt(np.sum((c_diversity - k_diversity) ** 2))

print("distance ",dist)
```
